```python
# load libraries
%matplotlib inline
import pandas as pd
import numpy as np
import time
import os
```

```python
#Import a csv with just the course names and numbers - modified in Excel
df = pd.read_csv("IDM_course_groups.csv", header=0)
#Show the first 5 rows to verify
df.head()
```

```python
#Initialize a new dataframe with the relevant column names
init_data = [['course','group',0]]
output_df = pd.DataFrame(init_data, columns = ['Course','Group','Size'])
#See what it looks like to verify
output_df.head()
```

```python
#Output dataframe index tracker
k=0

#Iterate through all rows of the original dataframe
for i in df.index:
    #Save the course number and course name in temporary variabels
    group = df.at[i,'COURSE']
    course = df.at[i,'COURSE_NAME']
    
    #Iterate through the output df and check for repeates
    repeat_flag = False
    for j in output_df.index:
        #If you find a repeat, increment the size by one but don't do anything else
        if output_df.at[j,'Course'] == course:
            output_df.at[j,'Size'] += 1
            repeat_flag = True
            break;
            
    #If you don't find a repeat, add it to the output Df
    if repeat_flag == False:
        temp_data = [[course, group, 1]]
        temp_df = pd.DataFrame(temp_data, columns = ['Course', 'Group', 'Size'])
        output_df.loc[k] = temp_df.loc[0]
        #Move to the next row in the output DF
        k += 1

#Show the output DF to verify
output_df.head()
```

```python
#Try sorting it to get rid of the annoying thing where it treats numbers as strings, eg. 10 comes before 2, but it doesn't work
output_df.sort_values(by=['Group'])
```

```python
#save the final dataframe to a CSV file in the local directory
output_df.to_csv(r'IDM_nodes2.csv', index = False)
```
